```python
class Mylist(list):   #this is inheritance. I am inheriting an inbuilt class called List, which acts as my parent class.
  def __repr__(self):
    s = "< "
    for i in self:
      s += str(i) + " | "
    s += " >"
    return s
```

## Inheritance

Class variables are defined outside of any method, but within the class definition. They are shared by all instances of the class.
Instance variables are defined within the __init__ method and are specific to each instance of the class.

```python
class Car:
    # Class variable
    wheels = 4

    def __init__(self, color, model):
        # Instance variables
        self.color = color
        self.model = model

car1 = Car("red", "Model A")
car2 = Car("blue", "Model B")

print(car1.wheels)   # Output: 4
print(car2.wheels)   # Output: 4

print(car1.color)   # Output: red
print(car2.color)   # Output: blue
```
